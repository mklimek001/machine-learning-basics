## Feature selection

```python
from scipy.io import loadmat

data_raw = loadmat('Zadanie5-leukemia.mat')
X_raw = data_raw['X']
y = data_raw['Y']

data_raw
```

```python
import pandas as pd

X_df = pd.DataFrame(X_raw)
X_df
```

### Highest variance

```python
variances = X_df.var()
print(variances)
```

```python
def find_variance_lower_than(variances, P=1):
  columns = []
  for i in range(len(variances)):
    if variances[i] < P:
      columns.append(i)
  
  return len(columns), columns

n, cols_to_del = find_variance_lower_than(variances)
print(n)
```

```python
for i in [1, 1.25, 1.5, 1.75, 2, 2.5, 3]:
  n, cols_to_del = find_variance_lower_than(variances, i)
  print("variance = ", i, ' ---> ', n, 'features to delete')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

n, cols_to_del = find_variance_lower_than(variances, 2)
X_var_2 = X_df.drop(cols_to_del, axis=1)
X_var_2
```

```python
n, cols_to_del = find_variance_lower_than(variances, 3)
X_var_3 = X_df.drop(cols_to_del, axis=1)
X_var_3
```

```python
indexes = list(X_df.index)
```

```python
from sklearn.model_selection import train_test_split

i_train, i_test = train_test_split(indexes, test_size=0.25)
```

```python
print(i_train)
print(i_test)
```

```python
y_arr = []
for i in range(len(y)):
  y_arr.append(y[i][0])

y_pd = pd.Series(y_arr).replace(-1, 0)
y_pd
```

```python
X_df.index == X_var_2.index
```

```python
def split_with_indexes(i_train, i_test, data):
  return (data.iloc[i_train], data.iloc[i_test])

X_train, X_test = split_with_indexes(i_train, i_test, X_df)
y_train, y_test = split_with_indexes(i_train, i_test, y_pd)
X_train_var_2, X_test_var_2 = split_with_indexes(i_train, i_test, X_var_2)
X_train_var_3, X_test_var_3 = split_with_indexes(i_train, i_test, X_var_3)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression(random_state=42)
lr_clf.fit(X_train, y_train)

y_pred_all = lr_clf.predict(X_test)
y_pred_all
```

```python
lr_clf = LogisticRegression(random_state=42)
lr_clf.fit(X_train_var_2, y_train)

y_pred_var_2 = lr_clf.predict(X_test_var_2)


lr_clf = LogisticRegression(random_state=42)
lr_clf.fit(X_train_var_3, y_train)

y_pred_var_3 = lr_clf.predict(X_test_var_3)
```

```python
lr_comparison = pd.DataFrame({
    'true': y_test,
    'all': y_pred_all,
    'var_2': y_pred_var_2,
    'var_3': y_pred_var_3
})

lr_comparison
```

```python
from sklearn.metrics import accuracy_score, auc, roc_curve

def calc_metrics(y_true, y_pred):
  fpr, tpr, thresholds = roc_curve(y_true, y_pred, pos_label=2)
  return {'auc': auc(fpr, tpr), 'accurracy': accuracy_score(y_true, y_pred)} 


print('Logistics regression: ')
print('All parameters: ', calc_metrics(lr_comparison['true'], lr_comparison['all']))
print('Parameters with variance > 2: ', calc_metrics(lr_comparison['true'], lr_comparison['var_2']))
print('Parameters with variance > 3: ', calc_metrics(lr_comparison['true'], lr_comparison['var_3']))
```

Przy tak małym zbiorze próbek niemożliwe jest wyliczenie metryki AUC. Dokładność, niezależnie od ilości wybranych za pomocą wariancji parametrów, wynosi 1.

```python
from sklearn.ensemble import RandomForestClassifier

rfc_clf = RandomForestClassifier(random_state=42)
rfc_clf.fit(X_train, y_train)

y_pred_all = rfc_clf.predict(X_test)

rfc_clf = RandomForestClassifier(random_state=42)
rfc_clf.fit(X_train_var_2, y_train)

y_pred_var_2 = rfc_clf.predict(X_test_var_2)


rfc_clf = RandomForestClassifier(random_state=42)
rfc_clf.fit(X_train_var_3, y_train)

y_pred_var_3 = rfc_clf.predict(X_test_var_3)
```

```python
rfc_comparison = pd.DataFrame({
    'true': y_test,
    'all': y_pred_all,
    'var_2': y_pred_var_2,
    'var_3': y_pred_var_3
})

rfc_comparison
```

```python
print('Random forrest: ')
print('All parameters: ', calc_metrics(rfc_comparison['true'], rfc_comparison['all']))
print('Parameters with variance > 2: ', calc_metrics(rfc_comparison['true'], rfc_comparison['var_2']))
print('Parameters with variance > 3: ', calc_metrics(rfc_comparison['true'], rfc_comparison['var_3']))
```

Przy zastosowaniu drzew losowych, predykcje z wykorzystaniem mniejszej liczby cech okazują się nawet bardziej dokładne. Zbyt mała liczba próbek nie pozwala na polczenia AUC.

### Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE

def get_lr_fs(X, y, f_num):
  clf = LogisticRegression(random_state=42)
  selector = RFE(clf, n_features_to_select=f_num, step=1)
  selector.fit(X, y)
  return selector


def get_rfc_fs(X, y, f_num):
  # clf = RandomForestClassifier(random_state=42)
  clf = RandomForestClassifier(random_state=42, n_estimators=10)
  selector = RFE(clf, n_features_to_select=f_num, step=1)
  selector.fit(X, y)
  return selector
```

```python
len(y_train)//3
```

```python
lr_selector_18 = get_lr_fs(X_train, y_train, 18)
```

```python
lr_18 = lr_selector_18.predict(X_test)
```

```python
rfc_selector_18 = get_rfc_fs(X_train, y_train, 18)
```

```python
rfc_18 = rfc_selector_18.predict(X_test)
```

```python
rfc_all = RandomForestClassifier(random_state=42)
rfc_all.fit(X_train, y_train)

lr_all = LogisticRegression(random_state=42)
lr_all.fit(X_train, y_train)
```

```python
rfe_comparison = pd.DataFrame({
    'true': y_test,
    'rfc_all' : rfc_all.predict(X_test),
    'rfc_18' : rfc_18,
    'lr_all' : lr_all.predict(X_test),
    'lr_18' : lr_18
})

rfe_comparison
```

```python
print('Recursive feature elimination')
print('\n Random forrest: ')
print('All parameters: ', calc_metrics(rfe_comparison['true'], rfe_comparison['rfc_all']))
print('18 parameters choosen with RFE: ', calc_metrics(rfe_comparison['true'], rfe_comparison['rfc_18']))

print('\n Logistics regression: ')
print('All parameters: ', calc_metrics(rfe_comparison['true'], rfe_comparison['lr_all']))
print('18 parameters choosen with RFE: ', calc_metrics(rfe_comparison['true'], rfe_comparison['lr_18']))
```

W przypadku drzew losowych, zmniejszona do 18 liczba cech nie pozwoliła na uzyskanie idealnego dopasowania w przypadku drzew losowych. Dla regresji logistycznej wyniki są w obydwu przypadkach tak samo dobre.   
Wybranie cech z wykorzystaniem rfe trwa jednak, nawet przy tak prostych klasyfikatorach bardzo długo (z wykorzystaniem google colab ~15 minut w każdym przypadku).

Scoring with cross validation

### Built-in regularization methods

```python
lr_reg_l1 = LogisticRegression(random_state=42, solver='saga', penalty='l1', max_iter=10000)
lr_reg_l1.fit(X_train, y_train)

y_lr_l1 = lr_reg_l1.predict(X_test)
```

```python
print('Logistics regression l1 regularization: ', calc_metrics(y_test, y_lr_l1))
```

W przypadku regularyzacji L1 dla domyślnych parametrów wyniki są również tak  dobre, jak we wcześniejszych przypadkach.

```python
from sklearn.feature_selection import SelectFromModel

selector = SelectFromModel(RandomForestClassifier(random_state=42, n_estimators=10))
selector.fit(X_train, y_train)

selected_features = X_train.columns[(selector.get_support())]
print(selected_features)
print(len(selected_features))
```

```python
X_t_train = selector.transform(X_train)
X_t_test = selector.transform(X_test)
```

```python
rfc_transformed = RandomForestClassifier(random_state=42, n_estimators=10)
rfc_transformed.fit(X_t_train, y_train)

y_transformed = rfc_transformed.predict(X_t_test)
print('Features selection from RFC model: ', calc_metrics(y_test, y_transformed))
```
