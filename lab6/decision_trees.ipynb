### Comparison

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier 
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
```

```python
NUM_TRIALS = 5

def calc_score(estm, p_grid, X, y, scoring):
    nested_scores = np.zeros(3 * NUM_TRIALS)
    best_params = []

    for i in range(NUM_TRIALS):
        inner_cv = KFold(n_splits=5, shuffle=True, random_state=i)
        outer_cv = KFold(n_splits=3, shuffle=True, random_state=i)

        clf = GridSearchCV(estimator=estm, param_grid=p_grid, cv=inner_cv, scoring=scoring)
        nested_score = cross_val_score(clf, X=X, y=y, cv=outer_cv, scoring=scoring)
        print(nested_score)
        nested_scores[3*i:3*i+3] = nested_score

        clf.fit(X, y)
        best_params.append(clf.best_params_)

        mean_nested_score = np.mean(nested_scores)
    
    return np.mean(nested_scores), np.std(nested_scores), best_params
```

```python
scoring_functions = ['accuracy', 'recall', 'precision']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

dt_clf = DecisionTreeClassifier(random_state=1)

dt_pipe = Pipeline([('std', StandardScaler()),
                  ('dt_clf', dt_clf)])

param_grid_dt = [{'dt_clf__min_samples_leaf': [5, 10, 20],
                'dt_clf__max_depth': [2, 5, 10]}]
```

```python
print('DecisionTreeClassifier')
for func in scoring_functions:
    mean_score, std, best_params = calc_score(dt_pipe, param_grid_dt, X, y, func)
    print('metrics - ', func, ' (mean) : ', mean_score)
    print('standard deviation : ', std)
    print('best parameteres : ', best_params)
```

```python
rf_clf = RandomForestClassifier(random_state=1)
et_clf = ExtraTreesClassifier(random_state=1)

rf_pipe = Pipeline([('std', StandardScaler()),
                  ('rf_clf', rf_clf)])

et_pipe = Pipeline([('std', StandardScaler()),
                  ('et_clf', et_clf)])


param_grid_rf = [{'rf_clf__n_estimators': [50, 100],
                'rf_clf__min_samples_leaf': [5, 10],
                'rf_clf__max_features': ['sqrt', 'log2']}]

param_grid_et = [{'et_clf__n_estimators': [50, 100],
                'et_clf__min_samples_split': [2, 5],
                'et_clf__max_features': ['sqrt', 'log2']}]
```

```python
print('RandomForestClassifier')
for func in scoring_functions:
    mean_score, std, best_params = calc_score(rf_pipe, param_grid_rf, X, y, func)
    print('metrics - ', func, ' (mean) : ', mean_score)
    print('standard deviation : ', std)
    print('best parameteres : ', best_params)

print('\n\n')

print('ExtraTreesClassifier')
for func in scoring_functions:
    mean_score, std, best_params = calc_score(et_pipe, param_grid_et, X, y, func)
    print('metrics - ', func, ' (mean) : ', mean_score)
    print('standard deviation : ', std)
    print('best parameteres : ', best_params)
```

W przypadku pojedynczego drzewa losowego żadna z metryk nie wypada zauważalnie gorzej od bardziej złożonych klasyfikatorów. Wybór parametrów (maksymalnej głębokości drzewa oraz minimalnego rozmiaru liścia) nie jest jednoznaczny. Jako maksymalna głębokość zdecydowanie najczęściej wskazywane jest 10. Algorytm działa szybciej, a jego struktura jest na tyle przejrzysta, że można ją łatwo przedstawić na grafie. 

Najlepszym klasyfikatorem okazuje się ten wykorzystujący lasy losowe. W ich przypadku wskazania wszystkich precission i accuracy są najwyższe, ale różnice nie są przesadnie duże. Najczęśćiej wybierany jest algorytm wykorzystujący 100 drzew i liście składające się z minimum 5 obiektów. Maksymalna ilość cech zależy od wybranej metryki - dla accuracy i recall najczęściej wybierana jest wartość pierwiastka kwadrtowego, a w przypadku precision logarytmu.

Algorytm extra trees działa szybciej niż lasy losowe i zapewnia najwyższy recall ze wszystkich sprawdzonych algorytmów. W wyborze cech najczęśćiej wskazywany jest pierwiastek jako maksymalna liczba cech, minimalna liczba próbek do podziału to najczęściej 5, a liczba drzew to 100. Algorytm ten działa też szybciej niż lasy losowe.



### Feature importances

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier 
import matplotlib.pyplot as plt
```

```python
dt_clf = DecisionTreeClassifier(random_state=1, min_samples_leaf=5, max_depth=10)

rf_clf = RandomForestClassifier(random_state=1, n_estimators=100, min_samples_leaf=5, max_features='sqrt')

et_clf = ExtraTreesClassifier(random_state=1, n_estimators=100, min_samples_split=5, max_features='sqrt')
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X))
X_scaled.columns = X.columns
X_scaled.head()
```

```python
def feature_importances_graph(clf, X, y, num=20):
    clf.fit(X, y)
    sorted_fi = clf.feature_importances_.argsort()[-num:]
    plt.figure(figsize=(8, num//3))
    plt.barh(X.columns[sorted_fi], clf.feature_importances_[sorted_fi], height=0.7)
    plt.xlabel("Feature Importance")
```

```python
feature_importances_graph(dt_clf, X_scaled, y)
```

```python
feature_importances_graph(rf_clf, X_scaled, y, 40)
```

```python
feature_importances_graph(et_clf, X_scaled, y, 50)
```

W każdym przypadku najbardziej istotną cechą okazuje się '5_Marriec-civ-spouse'. Istotne są też w każdym przypadku numeryczne cechy, które nie powstały w wyniku procedury 'get_dummies' przeprowadzonej na pierwotnych cechach kategorycznych.

Największe znaczenie pojedyncze cechy mają w przypadku drzewa decyzyjnego. Na końcowy rezultat w tym przypadku wpływ ma jedynie około 15 z 96 cech. Dla lasów losowych i algorytmu ExtraTrees ważnych jest znacznie więcej, nawet ponad 40, 50 cech. Pojedyncze z nich nie są jednak tak istotne.

W przypadku drzew decyzyjnych możemy zaprezentować wyniki ich działania w formie wygodnego do zrozumienia grafu:

```python
from sklearn import tree

dt_clf = DecisionTreeClassifier(random_state=1, min_samples_leaf=20, max_depth=5)
dt_clf.fit(X_scaled, y)

plt.figure(figsize=(20, 10))
tree.plot_tree(dt_clf,
               feature_names=X.columns,
               filled=True,
               fontsize=8)
```
